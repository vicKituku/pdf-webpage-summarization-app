```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = "https://kenyanwallstreet.com/"
page_url = "https://kenyanwallstreet.com/undersea-cable-cuts-disrupt-internet-in-several-african-countries/"
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36'
}

response = requests.get(page_url,headers=headers)
business_soup = BeautifulSoup(response.text, "html.parser")
business_soup.find('h1', attrs={'class': 'jeg_post_title'}).text
business_soup.find('div', attrs={'class': 'content-inner'}).text
```

```python
article_links_url = "https://kenyanwallstreet.com/category/african-news/"
articles_response = requests.get(article_links_url, headers=headers)
articles_soup  = BeautifulSoup(articles_response.text, "html.parser")
```

```python
links = articles_soup.find_all('h3', attrs={'class': 'jeg_post_title'})
for link in links:
    print(link.find('a')['href'])
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

links = []

# Initialize Chrome webdriver
driver = webdriver.Chrome()

# Open the website
driver.get('https://kenyanwallstreet.com/category/african-news/')

# Define a function to click the "load more" button
def click_load_more():
    try:
        load_more_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '/html/body/div[2]/div[5]/div/div[1]/div[2]/div/div[4]/div[1]/div/div[2]/div/div[2]/div[2]/a'))
        )
        load_more_button.click()
        return True
    except:
        return False
#Scrape links
while True:
    # Get links from the current page
    page_links = driver.find_elements(By.XPATH, '//h3[@class="jeg_post_title"]/a')
    links.extend([link.get_attribute('href') for link in page_links])

    # Click the "load more" button
    if not click_load_more():
        break
driver.quit()
```

```python
for link in links:
    response = requests.get(link, headers=headers)
    news_soup = BeautifulSoup(response.text, "html.parser")
    title = news_soup.find('h1', attrs={'class': 'jeg_post_title'}).text
    body = news_soup.find('div', attrs={'class': 'content-inner'}).text
    print(title, body)
    
```
